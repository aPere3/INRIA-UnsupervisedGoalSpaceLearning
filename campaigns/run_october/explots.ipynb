```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.misc
%matplotlib inline
```

```python
data = scipy.misc.imread('test.png', mode='RGB')
```

```python
a = np.load("results/<class 'embeddings.AutoEncoderEmbedding'>-207889d6b7f711e7ae942c4d54448326.npy")
h, _, _ = np.histogram2d(a[:,0], a[:,1],bins=10, range=[(-1,1), (-1,1)])
h[h!=0.] = 1.
plt.figure(figsize=(7,7))
plt.xticks(np.arange(-1,1,2./10.))
plt.yticks(np.arange(-1,1,2./10.))
plt.grid(alpha=.2)
plt.scatter(a[:,0], a[:,1],marker=".",alpha=.5)
plt.imshow(data, extent = (-1.17,1,-1.1,1))
plt.imshow(np.rot90(h), alpha=.4, extent = (-1,1,-1,1), cmap="GnBu")
plt.axhline(0., c="black", alpha=.5)
plt.axvline(0., c="black", alpha=.5)
plt.title("Explored Cells with architectures RGE-AE")
plt.savefig("results/explots_rge_ae.pdf",bbox_inches='tight')
```
